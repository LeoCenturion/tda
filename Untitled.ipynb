```python
import pandas as pd
import random as rd
from stringMatching import naive 
from stringMatching import karpRabin
from stringMatching import jaccardKarpRabin
import os
import random
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("resultados.csv")
df =df[["text", "len_text", "pattern", "len_pattern", "algorithm", "time", "matches"]]
```

```python
df.time=df.time.map(lambda x: int(x))
df.len_text=df.len_text.map(lambda x: int(x))
df.len_pattern=df.len_pattern.map(lambda x: int(x))
```

### Tiempo vs longitud del texto: algoritmo ingenuo

```python
f, axarr = plt.subplots(3, sharex=True)
axarr[0].scatter(y=df[df.algorithm=="naive"]["time"],x=df["len_text"][df.algorithm=="naive"])
axarr[1].scatter(y=df["time"][df.algorithm=="kr"],x=df["len_text"][df.algorithm=="kr"])
axarr[2].scatter(y=df[df.algorithm!="naive"][df.algorithm!="kr"]["time"].values,\
            x=df[df.algorithm!="naive"][df.algorithm!="kr"]["len_text"].values)
f.subplots_adjust(hspace=0.3)
f.set_figheight(20)
f.set_figwidth(7)
plt.show()
```

### Tiempo vs longitud del patron: Karp Rabin

```python
f, axarr = plt.subplots(3, sharex=True)
axarr[0].scatter(y=df[df.algorithm=="naive"]["time"],x=df["len_pattern"][df.algorithm=="naive"])
axarr[1].scatter(y=df["time"][df.algorithm=="kr"],x=df["len_pattern"][df.algorithm=="kr"])
axarr[2].scatter(y=df[df.algorithm!="naive"][df.algorithm!="kr"]["time"].values,\
            x=df[df.algorithm!="naive"][df.algorithm!="kr"]["len_pattern"].values)
f.subplots_adjust(hspace=0.3)
f.set_figheight(20)
f.set_figwidth(7)
plt.show()
```

### Tiempo vs longitud del texto por la longitud del patron: Jaccard Karp Rabin

```python


df["text*pattern"] = df[["len_text","len_pattern"]].apply(lambda row: int(row[0])*int(row[1]),axis=1)
f, axarr = plt.subplots(3, sharex=True)
axarr[0].scatter(y=df[df.algorithm=="naive"]["time"],x=df["text*pattern"][df.algorithm=="naive"])
axarr[1].scatter(y=df["time"][df.algorithm=="kr"],x=df["text*pattern"][df.algorithm=="kr"])
axarr[2].scatter(y=df[df.algorithm!="naive"][df.algorithm!="kr"]["time"].values,\
            x=df[df.algorithm!="naive"][df.algorithm!="kr"]["text*pattern"].values)
f.subplots_adjust(hspace=0.3)
f.set_figheight(20)
f.set_figwidth(7)
plt.show()
```

```python
f = lambda row: (row[6]+1)/(df[df.text==row[0]][df.pattern==row[2]][df.algorithm=="naive"]["matches"].values[0]+1)
df["falsePositivesRatio"] = df.apply(f,axis=1)
```
